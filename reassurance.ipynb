### **1. THÃ‰ORIE FONDAMENTALE DE LA RÃ‰ASSURANCE**

```python
def theories_fondamentales():
    """
    Concepts thÃ©oriques essentiels pour maÃ®triser la rÃ©assurance
    """
    
    st.markdown("""
    ### ğŸ§  **THÃ‰ORIE : LES FONDEMENTS Ã‰CONOMIQUES**
    
    #### **A. Principe de Mutualisation AvancÃ©e**
    """)
    
    st.latex(r'''
    \text{VariabilitÃ© du portefeuille} = \frac{\sigma_{\text{risque}}}{\sqrt{n}} \times \sqrt{1 + \rho(n-1)}
    ''')
    
    st.markdown("""
    OÃ¹ :
    - Ïƒ_risque : Ã‰cart-type d'un risque individuel
    - n : Nombre de risques dans le portefeuille  
    - Ï : CorrÃ©lation moyenne entre les risques
    
    #### **B. ThÃ©orie de l'UtilitÃ© et Aversion au Risque**
    
    La rÃ©assurance s'explique par la **fonction d'utilitÃ© concave** des assureurs :
    """)
    
    st.latex(r'''
    U(W) = \frac{W^{1-\gamma}}{1-\gamma} \quad \text{avec } \gamma > 0
    ''')
    
    st.markdown("""
    OÃ¹ Î³ mesure le degrÃ© d'aversion au risque. Plus Î³ est Ã©levÃ©, plus l'assureur est prÃªt Ã  payer pour transfÃ©rer le risque.
    
    #### **C. Value at Risk (VaR) et Tail Value at Risk (TVaR)**
    
    Mesures de risque utilisÃ©es pour dimensionner les programmes de rÃ©assurance :
    """)
    
    st.latex(r'''
    \begin{align*}
    \text{VaR}_{\alpha}(X) &= \inf\{x \in \mathbb{R} : F_X(x) \geq \alpha\} \\
    \text{TVaR}_{\alpha}(X) &= \mathbb{E}[X | X > \text{VaR}_{\alpha}(X)]
    \end{align*}
    ''')
```

### **2. COMPTABILITÃ‰ TECHNIQUE AVANCÃ‰E**

```python
def comptabilite_technique_avancee():
    """
    Principes comptables techniques spÃ©cifiques Ã  la rÃ©assurance
    """
    
    st.markdown("""
    ### ğŸ“Š **COMPTABILITÃ‰ TECHNIQUE : PRINCIPES COMPTABLES SPÃ‰CIFIQUES**
    
    #### **A. Provisions Techniques (Article R. 343-1 Code des Assurances)**
    """)
    
    provisions = {
        'Provision': ['Provision pour Sinistres Ã  Payer (PSAP)', 'Provision pour Primes Non Acquises (PPNA)', 
                     'Provision pour AlÃ©as (PRA)', 'Provision pour Equalisation'],
        'Objet': ['Sinistres survenus non rÃ©glÃ©s', 'Primes relatives Ã  la pÃ©riode de couverture future',
                 'Risques en cours non encore survenus', 'Lissage des rÃ©sultats techniques'],
        'MÃ©thode de Calcul': ['MÃ©thode Chain Ladder', 'MÃ©thode Pro-rata Temporis', 'Approche actuarielle', 'Moyenne mobile sur 7 ans'],
        'Comptes': ['Compte 49', 'Compte 48', 'Compte 495', 'Compte 149']
    }
    
    st.dataframe(pd.DataFrame(provisions))
    
    st.markdown("""
    #### **B. MÃ©thode Chain Ladder pour la PSAP**
    
    La mÃ©thode Chain Ladder est la plus utilisÃ©e pour estimer les sinistres Ã  payer :
    """)
    
    st.latex(r'''
    \hat{C}_{i,k} = C_{i,I-i} \times \hat{\lambda}_{I-i} \times \cdots \times \hat{\lambda}_{k-1}
    ''')
    
    st.markdown("""
    OÃ¹ :
    - C_i,k : Sinistres cumulÃ©s pour l'annÃ©e de survenance i Ã  la fin du dÃ©veloppement k
    - Î»_j : Facteur de dÃ©veloppement entre j et j+1
    
    #### **C. Comptabilisation des Cessions et Acceptations**
    
    **Au bilan de la cÃ©dante :**
    - Compte 61 : Cessions en rÃ©assurance (crÃ©diteur)
    - Compte 62 : Acceptations en rÃ©assurance (dÃ©biteur)
    
    **En comptabilitÃ© d'engagement :**
    - Les primes cÃ©dÃ©es sont comptabilisÃ©es en charges
    - Les sinistres cÃ©dÃ©s sont comptabilisÃ©s en produits
    """)
```

### **3. THÃ‰ORIE DES TRAITÃ‰S PROPORTIONNELS**

```python
def theorie_traites_proportionnels():
    """
    ThÃ©orie mathÃ©matique approfondie des traitÃ©s proportionnels
    """
    
    st.markdown("""
    ### ğŸ“ˆ **THÃ‰ORIE DES TRAITÃ‰S PROPORTIONNELS**
    
    #### **A. ModÃ¨le MathÃ©matique du Quota-Share**
    
    Soit un portefeuille de N risques indÃ©pendants. Pour chaque risque i :
    """)
    
    st.latex(r'''
    \begin{align*}
    P_i &: \text{Prime directe du risque } i \\
    S_i &: \text{Sinistre du risque } i \\
    \alpha &: \text{Taux de cession} \\
    C &: \text{Taux de commission}
    \end{align*}
    ''')
    
    st.markdown("""
    **RÃ©sultats pour la cÃ©dante :**
    """)
    
    st.latex(r'''
    \begin{align*}
    \text{Prime nette} &= \sum_{i=1}^N P_i \times (1 - \alpha) \\
    \text{Sinistre net} &= \sum_{i=1}^N S_i \times (1 - \alpha) \\
    \text{Commission reÃ§ue} &= \sum_{i=1}^N P_i \times \alpha \times C
    \end{align*}
    ''')
    
    st.markdown("""
    #### **B. Optimisation du Taux de Cession**
    
    Le taux de cession optimal Î±* maximise l'utilitÃ© espÃ©rÃ©e :
    """)
    
    st.latex(r'''
    \alpha^* = \arg\max_{\alpha} \mathbb{E}[U(W - S_{\text{net}} - P_{\text{net}} + \text{Commission})]
    ''')
    
    st.markdown("""
    #### **C. Impact sur le Capital Requis**
    
    Sous SolvabilitÃ© II, la rÃ©assurance rÃ©duit le Capital Requis (SCR) :
    """)
    
    st.latex(r'''
    \text{SCR}_{\text{aprÃ¨s}} = \text{SCR}_{\text{avant}} \times (1 - \alpha) \times \text{Facteur de reconnaissance}
    ''')
```

### **4. MODÃˆLES DE TARIFICATION AVANCÃ‰S**

```python
def modeles_tarification_avances():
    """
    ModÃ¨les actuariels avancÃ©s pour la tarification en rÃ©assurance
    """
    
    st.markdown("""
    ### ğŸ’° **MODÃˆLES ACTUARIELS DE TARIFICATION**
    
    #### **A. Prime Pure selon la ThÃ©orie de la Ruine**
    
    La prime pure minimale qui garantit une probabilitÃ© de ruine Îµ :
    """)
    
    st.latex(r'''
    \pi_{\text{min}} = \mathbb{E}[S] + \kappa \times \sigma(S) \times \sqrt{-2 \log \epsilon}
    ''')
    
    st.markdown("""
    #### **B. ModÃ¨le de Credibility Theory**
    
    Combinaison entre l'expÃ©rience propre et l'expÃ©rience collective :
    """)
    
    st.latex(r'''
    \hat{\mu} = Z \times \bar{X} + (1 - Z) \times M
    ''')
    
    st.markdown("""
    OÃ¹ :
    - Z = n/(n+K) : Facteur de crÃ©dibilitÃ©
    - K = ÏƒÂ²/Ï„Â² : Ratio de variance
    - M : Prime collective
    
    #### **C. ModÃ¨le de BÃ¼hlmann-Straub**
    
    Extension pour des donnÃ©es structurÃ©es :
    """)
    
    st.latex(r'''
    \hat{\mu}_i = Z_i \times \bar{X}_i + (1 - Z_i) \times \bar{X}
    '''
    )
    
    st.markdown("""
    #### **D. Tarification des TraitÃ©s Non-Proportionnels**
    
    Pour un traitÃ© Stop Loss avec prioritÃ© L et limite M :
    """)
    
    st.latex(r'''
    \text{Prime} = \mathbb{E}[\min(\max(S - L, 0), M)] = \int_L^{L+M} (s - L) f_S(s) ds + M \times (1 - F_S(L+M))
    ''')
```

```python
def solvabilite_ii_avance():
    """
    Principes avancÃ©s de SolvabilitÃ© II et gestion du capital
    """
    
    st.markdown("""
    ### ğŸ›¡ï¸ **SOLVABILITÃ‰ II : CADRE AVANCÃ‰**
    
    #### **A. Architecture des Trois Piliers**
    """)
    
    piliers = {
        'Pilier': ['Pilier I - Exigences Quantitatives', 'Pilier II - Exigences Qualitatives', 'Pilier III - Transparence'],
        'Composantes': [
            'SCR, MCR, Provisions Techniques, Actifs',
            'ORSA, Gouvernance, ContrÃ´le Interne, Risk Management',
            'Reporting, Disclosure, Transparence marchÃ©'
        ],
        'Textes RÃ©fÃ©rents': [
            'Directive 2009/138/CE, DÃ©lÃ©guÃ©s 2015/35',
            'Loi Sapin 2, Guidelines EIOPA',
            'Implementing Technical Standards'
        ]
    }
    
    st.dataframe(pd.DataFrame(piliers))
    
    st.markdown("""
    #### **B. Calcul du Solvency Capital Requirement (SCR)**
    
    Approche standard pour le module de souscription non-vie :
    """)
    
    st.latex(r'''
    \text{SCR}_{\text{non-vie}} = \sqrt{\sum_{i,j} \rho_{i,j} \times \text{SCR}_i \times \text{SCR}_j}
    ''')
    
    st.markdown("""
    OÃ¹ Ï_i,j sont les corrÃ©lations entre les sous-modules.
    
    #### **C. Effet de la RÃ©assurance sur le SCR**
    
    La rÃ©assurance rÃ©duit le SCR via deux mÃ©canismes :
    
    1. **Effet direct** : RÃ©duction des provisions techniques
    2. **Effet indirect** : RÃ©duction de la volatilitÃ© des sinistres
    
    **Reconnaissance de la rÃ©assurance :**
    """)
    
    st.latex(r'''
    \text{SCR}_{\text{aprÃ¨s}} = \text{SCR}_{\text{avant}} \times \left(1 - \frac{\text{Couvertures reconnues}}{\text{Exposition brute}}\right)
    ''")
```

### **6. RISQUES CATASTROPHIQUES ET MODÃ‰LISATION**

```python
def modelisation_catastrophes():
    """
    ModÃ©lisation avancÃ©e des risques catastrophiques
    """
    
    st.markdown("""
    ### ğŸŒªï¸ **MODÃ‰LISATION DES RISQUES CATASTROPHIQUES**
    
    #### **A. Processus de Poisson ComposÃ© pour les Catastrophes**
    
    ModÃ©lisation du nombre et de l'intensitÃ© des catastrophes :
    """)
    
    st.latex(r'''
    S(t) = \sum_{i=1}^{N(t)} X_i
    '''
    )
    
    st.markdown("""
    OÃ¹ :
    - N(t) ~ Poisson(Î»t) : Nombre de catastrophes
    - X_i : Montant du sinistre de la catastrophe i
    - X_i ~ LogNormal(Î¼, ÏƒÂ²) : Distribution des sinistres
    
    #### **B. Courbes de Dommage (Damage Ratios)**
    
    Pour un Ã©vÃ©nement d'intensitÃ© I Ã  la localisation L :
    """)
    
    st.latex(r'''
    \text{Dommage}(I, L) = \text{Exposition}(L) \times \text{Damage Ratio}(I) \times \text{Alerte}(L)
    ''')
    
    st.markdown("""
    #### **C. Value at Risk Catastrophe (Cat VaR)**
    
    Mesure du risque extrÃªme pour les programmes cat :
    """)
    
    st.latex(r'''
    \text{Cat VaR}_{\alpha} = \inf\{x : \mathbb{P}(S_{\text{cat}} > x) \leq 1 - \alpha\}
    ''')
    
    st.markdown("""
    #### **D. ProbabilitÃ© de DÃ©passement (Exceedance Probability)**
    
    Courbe EP fondamentale pour la tarification :
    """)
    
    st.latex(r'''
    \text{EP}(x) = \mathbb{P}(S_{\text{cat}} > x) = 1 - F_{S_{\text{cat}}}(x)
    ''')
```
